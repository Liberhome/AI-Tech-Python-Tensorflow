# 手写数字识别

```python
import keras
from keras import layers
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import keras.datasets.mnist as mnist
(train_image,train_label),(test_image,test_label)=mnist.load_data()
```

```python
train_image.shape
```

```python
plt.imshow(train_image[1])
```

```python
train_label
```

```python
model=keras.Sequential()
# flatten 是 否 展 开 输 入 图 像 ( 变 成 一 维 数 组 ) 如果为不做 输 入 图 像为 60000× 28 × 28 的三维数组，
#做了，则输入图像会保存为（60000，28*28）的二维数据
model.add(layers.Flatten())
model.add(layers.Dense(64,activation='relu'))#输入为28*28，输出64个单元的隐藏层 relu激活
model.add(layers.Dense(10,activation='softmax'))#最后的10个分类
```

```python
model.compile(optimizer='adam'
             ,loss='sparse_categorical_crossentropy'
             ,metrics=['acc'])
```

```python
model.fit(train_image,train_label,epochs=25,batch_size=128)
```

```python
#这里可以看到，在训练集上准确率97%+
```

```python
#我们来看看在测试集上如何呢
model.evaluate(test_image,test_label)
```

```python
#这里可以看到，在测试集上准确率95%+
```

```python
#做一下预测
model.predict(test_image[:10])#预测前面10张图片
```

```python
#预测前面10张图片
np.argmax(model.predict(test_image[:10]),axis=1)
```

```python
#看一下实际前10张
test_label[:10]
```

模型的优化 

```python
model=keras.Sequential()
model.add(layers.Flatten())
#多搞几个隐藏层
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))#最后的10个分类
```

```python
model.compile(optimizer='adam'
             ,loss='sparse_categorical_crossentropy'
             ,metrics=['acc'])
```

```python
model.fit(train_image,train_label,epochs=25,batch_size=128,validation_data=(test_image,test_label))
```
